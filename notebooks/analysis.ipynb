```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv('../data/tweets_with_labels.csv')
print(f'Loaded {len(df)} tweets')
df.head()
```

## Sentiment Distribution

```python
# Plot sentiment distribution
sentiment_counts = df['sentiment'].value_counts()
fig = px.bar(
    x=sentiment_counts.index,
    y=sentiment_counts.values,
    labels={'x': 'Sentiment', 'y': 'Count'},
    title='Tweet Sentiment Distribution',
    color=sentiment_counts.index,
    color_discrete_map={'positive': '#2ecc71', 'neutral': '#95a5a6', 'negative': '#e74c3c'}
)
fig.show()
```

## Cryptocurrency Distribution

```python
# Plot coin distribution
coin_counts = df['coin'].value_counts()
fig = px.bar(
    x=coin_counts.index,
    y=coin_counts.values,
    labels={'x': 'Cryptocurrency', 'y': 'Count'},
    title='Tweet Distribution by Cryptocurrency',
    color=coin_counts.index
)
fig.show()
```

## Sentiment by Cryptocurrency

```python
# Create crosstab and plot heatmap
crosstab = pd.crosstab(df['coin'], df['sentiment'])
fig = px.imshow(
    crosstab,
    text_auto=True,
    labels={'x': 'Sentiment', 'y': 'Cryptocurrency', 'color': 'Count'},
    title='Sentiment Distribution by Cryptocurrency',
    aspect='auto'
)
fig.show()
```

## Sample Tweets by Sentiment

```python
# Show sample tweets for each sentiment
for sentiment in ['positive', 'neutral', 'negative']:
    print(f'{sentiment.upper()} TWEETS')
    print('-' * 60)
    samples = df[df['sentiment'] == sentiment]['text'].head(3)
    for i, tweet in enumerate(samples, 1):
        print(f'{i}. {tweet[:150]}...')
    print()
```
